```python
# YOUR CODE HERE
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split 
from sklearn.metrics import log_loss
from sklearn.metrics import accuracy_score
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV
import random
```

```python
# YOUR CODE HERE
#preparing data for the model 
y=new_df['Happiness?']
X=new_df.drop(columns='Happiness?',axis=1)

#creating training and testing data 
X_train, X_test, y_train, y_test=train_test_split(X,y,test_size=.33,random_state=1234)
```

```python
#logistic regression model - taken from unit 4 assignment
#function to train and test a logistic regression model with a regularization factor of 1
def train_test_LR(X_train, y_train, X_test, y_test, c=1):
    model=LogisticRegression(C=c)
    model.fit(X_train,y_train)
    
    prob=model.predict_proba(X_test)
    predict=model.predict(X_test)
    
    acc=accuracy_score(predict,y_test)
    
    return acc

#test different values of c
cs=[10**i for i in range(-10,10)]
acc=[]
for i in cs:
    result=train_test_LR(X_train, y_train, X_test, y_test,i)
    acc.append(result)
    print(result)
    
#reformat to plot and plot the relationship between accuracy and c
cs_log10=np.log10(cs)
ax=sns.lineplot(x=cs_log10,y=acc,marker='o')
plt.title('Accuracy Scores vs. C')
ax.set_xlabel('C')
ax.set_ylabel('Accuracy Scores')
plt.show()

lr=LogisticRegression(C=1)
lr_cross_val_score=cross_val_score(lr,X,y,cv=5,scoring='accuracy').mean()
print(lr_cross_val_score)
```

After training the Logistic Regression model and testing different values of C for regularization, the plot shows that a value of 1 for C resulted in the highest accuracy score, which was 0.8378378378378378. With that hyperparameter, the cross validation score was slightly lower at 0.820624732563115.

```python
#random forest
#perform grid search to find best values of n-estimators and max depth
n_estims=[1,25,50,100,200,300,400,500]
max_depths=[1,8,16,32,64,128,256]
param_grid={'max_depth': max_depths, 'n_estimators': n_estims}
grid=GridSearchCV(RandomForestClassifier(),param_grid,cv=5,n_jobs=-1)
#read online that n_jobs can speed up computation time by using all CPUs
grid_search=grid.fit(X_train,y_train)

#find best values 
best_depth=grid_search.best_estimator_.max_depth
print(best_depth)
best_n=grid_search.best_estimator_.n_estimators
print(best_n)
```

```python
#new cell to get the accuracy score and cross validation score of a random forest model to avoid running the cell above again
#taken from lab 6 assignment 
#train RF with best hyperparameters 
rf=RandomForestClassifier(max_depth=128,n_estimators=100)
rf.fit(X_train,y_train)
rf_pred=rf.predict(X_test)

#get accuracy and cross validation score 
rf_acc=accuracy_score(y_test,rf_pred)
print(rf_acc)
rf_cross_val_score=cross_val_score(rf,X,y,cv=5,scoring='accuracy').mean()
print(rf_cross_val_score)
```

After training and testing the Random Forest model with the optimal hyperparameters, the accuracy score was 0.8796068796068796 while the cross validation score was 0.7889700799842008.

When comparing the two models, the Random Forest model had a higher accuracy score, but the Logistic Regression model had a higher cross validation score. Accuracy and cross validation scores both indicate how well the model performed on test data. However, cross validation is more accurate in how well a model can generalize to new, unseeen data because it evaluates performance over multiple folds of the data. Because the Logistic Regression model had a more consistent performance across the folds, it may be less prone to overfitting and more robust overall, even if its test accuracy was slightly lower.
